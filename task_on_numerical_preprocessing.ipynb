##### Missing Value Imputation (Mean, Median, Mode)

```python
Pregnancies_mean = df['Pregnancies'].fillna(df['Pregnancies'].mean())
```

```python
#here i used mean . the mean and median values are nearly same so i used mean
```

```python
Glucose_mean = df['Glucose'].fillna(df['Glucose'].mean())
```

```python
Glucose_mean.isna().sum()
```

```python
# for the Glucose i used mean because in between the mean and median there is only 3% difference so it will not much effect the model
```

```python
BloodPressure_mean = df['BloodPressure'].fillna(df['BloodPressure'].mean())
```

```python
BloodPressure_mean.isna().sum()
```

```python
# here the mean is used because the in between the mean and the median there the differ of 4 and there are many outlier in it 
```

```python
SkinThickness_median = df['SkinThickness'].fillna(df['SkinThickness'].median())
```

```python
SkinThickness_median.isna().sum()
```

```python
# here the median is used in between the min and 25 it is 30% incearse and so the mean and median the values also difference
```

```python
Insulin_median= df['Insulin'].fillna(df['Insulin'].median())
```

```python
Insulin_median.isna().sum()
```

```python
BMI_mean = df['BMI'].fillna(df['BMI'].mean())
```

```python
BMI_mean.isna().sum()
```

```python
#there is the no diff from mean and median
```

```python
DiabetesPedigreeFunction_median = df['DiabetesPedigreeFunction'].fillna(df['DiabetesPedigreeFunction'].median())
```

```python
DiabetesPedigreeFunction_median.isna().sum()
```

```python
Age_mean = df['Age'].fillna(df['Age'].mean())
```

```python
Age_mean
```

```python
Outcome_mean = df['Outcome'].fillna(df['Outcome'].mean())
```

```python
Outcome_mean.isna().sum()
```

```python
col = df.columns.to_list()
col
```

```python
df_processed = pd.DataFrame(columns=col)
df_processed
```

```python
df_processed['Age'] = Age_mean
```

```python
df_processed['Pregnancies'] = Pregnancies_mean
```

```python
df_processed['Glucose'] = Glucose_mean
```

```python
df_processed['BloodPressure'] = BloodPressure_mean
```

```python
df_processed['SkinThickness'] = SkinThickness_median
```

```python
df_processed['Insulin'] = Insulin_median
```

```python
df_processed['BMI'] = BMI_mean
```

```python
df_processed['DiabetesPedigreeFunction'] = DiabetesPedigreeFunction_median
```

```python
df_processed['Outcome'] = Outcome_mean
```

```python
df_processed
```

```python
df_processed.isna().sum()
```

```python
X = df_processed.drop('Outcome', axis=1)
y = df_processed['Outcome']
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test , y_train, y_test = train_test_split(X,y, train_size=0.8, random_state = 42)
X_train.shape, X_test.shape , y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().set_output(transform='pandas')
```

```python
sc
```

```python
par = sc.fit(X_train)
```

```python
par
```

```python
standard_scaled_df = sc.transform(X_train)
```

```python
standard_scaled_df.head()
```

```python
standard_scaled_df.to_csv('diabetes_standard_scaled.csv')
```

```python

num_cols = df_processed.select_dtypes(include=['int64', 'float64']).columns.tolist()
num_cols.remove("Outcome")   # don't scale target

def plot_original_vs_scaled(df_processed, scaled_df, title):
    num_cols = df_processed.columns

    plt.figure(figsize=(16, 10))

    for i, col in enumerate(num_cols):
        plt.subplot(3, 3, i + 1)
        plt.hist(df_processed[col], bins=30, alpha=0.5, label="Original")
        plt.hist(scaled_df[col], bins=30, alpha=0.5, label="Scaled")
        plt.title(col)
        plt.legend()

    plt.suptitle(title, fontsize=16)
    plt.tight_layout()
    plt.show()

# call the function
plot_original_vs_scaled(df_processed[num_cols], standard_scaled_df, "Original vs Standard Scaled")
```
